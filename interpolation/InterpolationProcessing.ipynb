```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style= 'whitegrid', font_scale=1.5)
```

```python
data = pd.read_csv('interpolation_errors.txt', header= None, dtype= np.float64)
h = [2 ** i for i in range(1, -5, -1)]
N = [3, 4, 5]
x_data = np.log(h)
```

```python
y_data3 = np.log(data.iloc[0, :-1])
y_data4 = np.log(data.iloc[1, :-1])
y_data5 = np.log(data.iloc[2, :-1])

p3 = np.polyfit(x_data, y_data3, 1)
y_3 = np.polyval(p3, x_data)
k3 = p3[0]

p4 = np.polyfit(x_data, y_data4, 1)
y_4 = np.polyval(p4, x_data)
k4 = p4[0]

p5 = np.polyfit(x_data, y_data5, 1)
y_5 = np.polyval(p5, x_data)
k5 = p5[0]


plt.figure(figsize= (16,10), dpi= 200)
plt.scatter(x_data, y_data3, c = 'red', label = f'N = 3: $\\tan \\alpha$ = {k3:.3f}')
plt.plot(x_data[:12], y_3, c = 'red', linestyle = '--')
plt.scatter(x_data, y_data4, c = 'green', label = f'N = 4: $\\tan \\alpha$ = {k4:.3f}')
plt.plot(x_data[:9], y_4, c = 'green', linestyle = '--')
plt.scatter(x_data, y_data5, c = 'purple', label = f'N = 5: $\\tan \\alpha$ = {k5:.3f}')
plt.plot(x_data[:7], y_5, c = 'purple', linestyle = '--')


plt.legend(loc = 'upper left')
plt.xlabel('$\ln h$')
plt.ylabel('$\ln err$')
plt.savefig(f'plot_interp')
```
