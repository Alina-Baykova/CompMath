```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

import matplotlib.ticker as ticker

sns.set_theme(style= 'whitegrid', font_scale=1.5)
```

```python
data = pd.read_csv('errKepler.txt', header= None, dtype= np.float64)
eccs = np.array([0.8, 0.5, 0.2, 0.1])
x_data = np.array(range(1, 7))
```

```python
y_data2 = np.log10(data.iloc[0, :-1])
y_data3 = np.log10(data.iloc[1, :-1])
y_data4 = np.log10(data.iloc[2, :-1])
y_data5 = np.log10(data.iloc[3, :-1])

fig = plt.figure(figsize= (16,10), dpi= 200)
ax = fig.add_subplot()

plt.scatter(x_data, y_data2, c = 'purple', label = f'e = {eccs[0]}')
plt.plot(x_data, y_data2, c = 'purple', linestyle = ':', alpha = 0.3)

plt.scatter(x_data, y_data3, c = 'brown', label = f'e = {eccs[1]}')
plt.plot(x_data, y_data3, c = 'brown', linestyle = ':', alpha = 0.3)

plt.scatter(x_data, y_data4, c = 'red', label = f'e = {eccs[2]}')
plt.plot(x_data, y_data4, c = 'red', linestyle = ':', alpha = 0.3)

plt.scatter(x_data, y_data5, c = 'blue', label = f'e = {eccs[3]}')
plt.plot(x_data, y_data5, c = 'blue', linestyle = ':', alpha = 0.3)

plt.title('      - \n     ')
plt.legend(loc = 'lower left')
plt.ylabel('$\lg err$')
plt.xlabel('$i$')
plt.savefig(f'plotKepler')
```
