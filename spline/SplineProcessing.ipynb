```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set_theme(style= 'whitegrid', font_scale=1.5)
dataNat = pd.read_csv('errNatural.txt', header= None, dtype= np.float64)
dataGen = pd.read_csv('errGeneral.txt', header= None, dtype= np.float64)
h = [5 * 2 ** i for i in range(6)]
x_data = np.log(h)
```

```python
y_data1 = np.log(dataNat.iloc[0, :-1])
y_data2 = np.log(dataGen.iloc[0, :-1])

p1 = np.polyfit(x_data, y_data1, 1)
y1 = np.polyval(p1, x_data)
k1 = p1[0]

p2 = np.polyfit(x_data, y_data2, 1)
y2 = np.polyval(p2, x_data)
k2 = p2[0]

plt.figure(figsize= (16,10), dpi= 200)
plt.scatter(x_data, y_data1, c = 'red')
plt.plot(x_data, y1, c = 'red', linestyle = '--', label = f' \n$\\tan \\alpha$ = {k1:.2f}')

plt.scatter(x_data, y_data2, c = 'green')
plt.plot(x_data, y2, c = 'green', linestyle = '--', label = f'    \n$\\tan \\alpha$ = {k2:.2f}')

plt.title(' $e^x$    $[0, 10]$\n      ')
plt.legend(loc = 'lower left')
plt.xlabel('$\ln h$')
plt.ylabel('$\ln err$')
plt.savefig(f'plotNatural')
```
