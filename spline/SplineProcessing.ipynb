```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set_theme(style= 'whitegrid', font_scale=1.5)
data = pd.read_csv('errNatural.txt', header= None, dtype= np.float64)
h = [5 * 2 ** i for i in range(6)]
x_data = np.log(h)
```

```python
y_data = np.log(data.iloc[0, :-1])

p = np.polyfit(x_data, y_data, 1)
y = np.polyval(p, x_data)
k = p[0]

plt.figure(figsize= (16,10), dpi= 200)
plt.scatter(x_data, y_data, c = 'red')
plt.plot(x_data, y, c = 'red', linestyle = '--', label = f' \n$\\tan \\alpha$ = {k:.2f}')

plt.title(' $e^x$    $[0, 10]$\n      ')
plt.legend(loc = 'lower left')
plt.xlabel('$\ln h$')
plt.ylabel('$\ln err$')
plt.savefig(f'plotNatural')
```
